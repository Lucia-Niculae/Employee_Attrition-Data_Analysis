- The closer the number is to 1 (or -1), the stronger the correlation.

#### <font color=blue> Age variable</font>

The higher the Age, the higher the MonthlyIncome. 
    
- #### A correlation coefficient of 0.51 between Age and JobLevel variables signifies a moderate positive correlation. 
That means that the higher the Age, the higher the JobLevel. 
    
- #### A correlation coefficient of 0.68 between Age and TotalWorkingYears variables signifies a moderate positive correlation. 
That means that the higher the Age, the higher the TotalWorkingYears variable.
    
#### <font color=blue> Job Level</font>

The more years worked, the higher the job level.   

- #### A correlation coefficient of 0.95 between JobLevel and MonthlyIncome variables signifies a very high positive correlation. The higher the job level, the higher the monthly salary.

#### <font color=blue> Percent Salary Hike</font>
